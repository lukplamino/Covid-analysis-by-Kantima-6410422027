```python
def cleansex(x):
  Female = ['หญิง', '\xa0หญิง', 'หญง','หญิิง', 'ญ', 'หณิง']
  Male = [ 'ช','ชาย','นาย']
  if x in Female:
    return 'Female'
  elif x in Male:
    return 'Male'
  else:
    return 'No data'

df['sex'] = df['sex'].apply(cleansex)
print(df.sex.unique())
df.groupby(["sex"]).count()
```
